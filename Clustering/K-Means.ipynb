```python
cols = X.columns
```

```python
from sklearn.preprocessing import MinMaxScaler

ms = MinMaxScaler()

X = ms.fit_transform(X)
```

```python
X = pd.DataFrame(X, columns=[cols])
```

```python
X.head()
```

```python
# Import library yang dibutuhkan
# from sklearn.cluster import KMeans
# from sklearn import datasets
import matplotlib.pyplot as plt

# # Baca dataset iris
# iris = datasets.load_iris()
# X = iris['data']

# Tentukan jumlah cluster yang paling optimal dengan menggunakan metode elbow
sse = {}
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, random_state=1).fit(X)
    sse[k] = kmeans.inertia_ # Inertia: sum of squared distances of samples to their closest cluster center

# Tampilkan hasil metode elbow dengan plot elbow
plt.figure()
plt.plot(list(sse.keys()), list(sse.values()))
plt.xlabel("Number of cluster")
plt.ylabel("Sum of Squared Errors")
plt.title("Elbow Plot")
plt.show()
```

```python
from sklearn.metrics import silhouette_score

silhouette_scores = []
for k in range(2, 10):
    kmeans = KMeans(n_clusters=k, random_state=1).fit(X)
    silhouette_scores.append(silhouette_score(X, kmeans.labels_))

# Tampilkan hasil metode silhouette score dengan plot silhouette score
plt.figure()
plt.plot(range(2, 10), silhouette_scores)
plt.xlabel("Number of cluster")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Plot")
plt.show()
```

**Use elbow method to find optimal number of clusters**

```python
from sklearn.cluster import KMeans
cs = []
for i in range(2, 10):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(X)
    cs.append(kmeans.inertia_)
plt.plot(range(2, 10), cs)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters optimal')
plt.ylabel('SSE')
plt.show()
plt.savefig('elbow-method.png')
```

```python
kmeans = KMeans(n_clusters=2, random_state=0)

kmeans.fit(X)

# check how many of the samples were correctly labeled
labels = kmeans.labels_

correct_labels = sum(y == labels)
print("Result: %d out of %d samples were correctly labeled." % (correct_labels, y.size))
print('Accuracy score: {0:0.2f}'. format(correct_labels/float(y.size)))
```

```python
from sklearn.metrics import adjusted_rand_score
from sklearn.metrics import silhouette_score
from sklearn.metrics import davies_bouldin_score
from sklearn.metrics import calinski_harabasz_score
# Buat objek KMeans dengan jumlah cluster sebanyak 3
kmeans = KMeans(n_clusters=2)

# Latih model KMeans dengan data wine
kmeans.fit(X)

# Dapatkan label yang diprediksi oleh model
y_pred = kmeans.predict(X)
# Hitung Indeks Rand dengan menggunakan y_pred sebagai label yang diprediksi dan y sebagai label sebenarnya
score_rand = adjusted_rand_score(y, y_pred)

# Hitung Indeks Silhouette dengan menggunakan y_pred sebagai label yang diprediksi dan X sebagai data
score_silhouette = silhouette_score(X, y_pred)

# Hitung Indeks Dunn dengan menggunakan y_pred sebagai label yang diprediksi dan X sebagai data
score_dunn = davies_bouldin_score(X, y_pred)

# Hitung Indeks Calinski-Harabasz dengan menggunakan y_pred sebagai label yang diprediksi dan X sebagai data
score_calinski = calinski_harabasz_score(X, y_pred)

# Cetak hasil evaluasi clustering
print(f"Nilai Indeks Adjust Rand      : {score_rand:.3f}")
print(f"Nilai indeks Silhouette       : {score_silhouette:.3f}")
print(f"Nilai Indeks DBS              : {score_dunn:.3f}")
print(f"Nilai Indeks Calinski-Harabasz: {score_calinski:.3f}")
```
