**Teknik Linspace**

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/nilai.csv')
data.head(9)
```

```python
bins = np.linspace(min(data['nilai tugas']), max(data['nilai tugas']), 3)
print(bins)
```

```python
kategori = ['Cukup','Baik', 'Baik Sekali']
```

```python
data['nilai_binned'] = pd.cut(data['nilai tugas'], bins=len(bins), labels=kategori, include_lowest=False)
```

```python
print (data)
```

```python
data['nilai_binned'] = pd.cut(data['nilai tugas'], bins=len(bins), labels=kategori, include_lowest=True)
```

```python
print (data)
```

**Teknik quantile**

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/nilai.csv', sep=',')
data.head(9)
```

```python
data['nilai tugas_binned3'] = pd.qcut(data['nilai tugas'], 3)
print (data)
```

```python
kategori = ['Cukup','Baik', 'Baik Sekali']
data['nilai tugas_binned3'] = pd.qcut(data['nilai tugas'], q=3, labels=kategori)
print (data)
```
