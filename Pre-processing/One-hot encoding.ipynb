```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/populasi.xlsx")
df
```

```python
y = OneHotEncoder(sparse=False)
y = y.fit_transform(df[['Kota']])
y
```

```python
df_y = pd.DataFrame(y)
df_y
```

```python
header = df['Kota'].sort_values()
df_y.columns = 'Kota_'+header
df_y
```

```python
df_join = df.join(df_y)
df_join
```

```python
labelencoder = LabelEncoder()
df_join['Tingkat_Populasi'] = labelencoder.fit_transform(df_join['Tingkat_Populasi'])
```
